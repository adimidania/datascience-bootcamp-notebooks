## 2. Lines

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
z = [3,6,9,12,15]

plt.plot(x,y)
plt.plot(x,z)
plt.legend(["y = 2x", "y= 3x"])
plt.show()
```

## 3. Scatter plots

```python
# Actually "fig" is the figure that we will be drawing on
# ax represents the graph that we are drawing, we can draw as many graphs as we want
# We can do some color's mapping to, by using the vector c

x = [1,2,3,4,5]
y = [6,7,8,9,10]
c = [1,1,2,3,4]

fig, ax = plt.subplots()
ax.scatter(x,y, c=c)
ax.set_title("Scatter Plot")
ax.set_xlabel("X")
ax.set_ylabel("Y")
```

## 4. Histograms

```python
data = [1,1,1,1,2,2,2,2,2,2,2,3,3]
fig, ax = plt.subplots()
ax.hist(data)
ax.set_title("Histogram")
ax.set_ylabel("Occurrence")
ax.set_xlabel("Data")
```

## 5. Bar charts

```python
x = [3, 1, 3, 12, 2, 4, 4] 
y = [3, 2, 1, 4, 5, 6, 7] 
plt.bar(x, y)
# Title to the plot
plt.title("Bar Chart")
# Adding the legends
plt.legend(["bar"])
plt.show()
```

## 6. Pie charts

```python
x = [1, 2, 3, 4] 
# this will explode the 1st wedge
# i.e. will separate the 1st wedge
# from the chart
e  =(0.1, 0, 0, 0)
# This will plot a simple pie chart
plt.pie(x, explode = e)
# Title to the plot
plt.title("Pie chart")
plt.show()
```

## 12. Line graph

```python
# Every year, the gas price changes in every country
# Let's see how it's changing in USA and Canada

# We fix the X axis and the Y axis
plt.plot(gas.Year, gas.USA, 'r.-', label='United States')
plt.plot(gas.Year, gas.Canada, 'b.-', label='Canada')

# Styling the graph
plt.title('Gas Prices over Time (in USD)', fontdict={'fontsize': 15})
plt.xlabel('Year')
plt.ylabel('US Dollars')

# I needed to add labels and colors to USA and Canada in order for everything to work
plt.legend()

# Displaying the Graph
plt.show()
```
